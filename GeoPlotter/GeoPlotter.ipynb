```python
import numpy as np
import folium as fo
import pandas as pd
%matplotlib inline
```

```python
map=fo.Map()
```

```python
map
```

```python
volcano=pd.read_csv("volcano.csv")
```

```python
lat=list(volcano["Latitude"])
lon=list(volcano["Longitude"])
name=list(volcano["Name"])
```

```python
vol=fo.FeatureGroup(name="My Map")
```

```python
for lat,lon,name in zip(lat,lon,name):
    vol.add_child(fo.Marker(location=[lat,lon] ,popup=name ,icon=fo.Icon(color="red")))
```

```python
map.add_child(vol)
```

```python
vol.add_child(fo.GeoJson(data=(open("world.json","r",encoding="utf-8-sig").read())))
```

```python
map.add_child(vol)
```

```python
population=pd.read_csv("us_cities_pop.csv")
```

```python
population.head()
```

```python
lat_pop=population["lat"]
lon_pop=population["lon"]
name_pop=population["name"]
pop_pop=population["pop"]
```

```python
po=fo.FeatureGroup(name="My Map")
```

```python
def mark(pop):
    if pop>35000:
        color="orange"
    elif pop>10000 and pop<=35000:
        color="blue"
    else:
        color="green"
    return color
```

```python
for lat,lon,name,pop in zip(lat_pop,lon_pop,name_pop,pop_pop):
    po.add_child(fo.Marker(location=[lat,lon],popup=[name,pop],icon=fo.Icon(color=mark(pop))))
```

```python
map.add_child(po)
```
